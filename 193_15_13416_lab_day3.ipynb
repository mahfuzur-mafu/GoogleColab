```python
slices = [4,2,1,3]
activity = ['sleeping','eating','working','playing']
cols = ['g','y','r','m']

plt.pie(slices,
        labels= activity,
        colors= cols , 
        startangle= 0, 
        shadow= True , 
        explode = (0,0,0,0), 
        autopct="%1.2f%%")

plt.title("Scatter plot")
plt.show()
```

```python
df = pd.read_csv('/content/iris.csv')
df
```

```python
fig, ax = plt.subplots()
plt.title("Iris dataset")
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

ax.scatter(df['sepal_length'], df['petal_width'])
fig.set_size_inches(10,8)
```

```python
colors = {'setosa' : 'r', 'versicolor' : 'g','virginica': 'b'}

fig, ax = plt.subplots()
plt.title("Iris dataset")
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

for i in range(len(df['sepal_length'])):
  ax.scatter(df['sepal_length'][i], df['sepal_width'][i],color=colors[df['species'][i]])
fig.set_size_inches(10,8)
```

```python
data = df['species'].value_counts()

name= data.index
count = data.values 

plt.title("Scores")
plt.xlabel('Name')
plt.ylabel('Total count')

plt.bar(name,count)
```

```python
slices = df['species'].value_counts()
activity = ['setosa' , 'versicolor' ,'virginica']
cols = ['g','b','r']

plt.pie(slices,
        labels= activity,
        colors= cols , 
        startangle= 90, 
        shadow= False , 
        explode = (0,.1,0), 
        autopct="%1.2f%%",
        radius=1.2)

plt.title("pie plot")
plt.show()
```
