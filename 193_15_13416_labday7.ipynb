```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/income2.csv")
```

```python
df
```

```python
df=df.drop('id',axis=1)
df=df.drop('name',axis=1,)
df
#df = df.drop(df['id'],df['name'],axis=1)
```

```python
plt.scatter(df['age'],df['income'])
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
y_pre= km.fit_predict(df[['age','income']])
y_pre
```

```python
df['cluster']=y_pre
df
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['age'],df1['income'],color='g')
plt.scatter(df2['age'],df2['income'],color='b')
plt.scatter(df3['age'],df3['income'],color='r')

plt.xlabel('Age')
plt.xlabel('Income')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
df['income']=scaler.fit_transform(df[['income']])
df['age']=scaler.fit_transform(df[['age']])
df
```

```python
km=KMeans(n_clusters=3)
y_pre= km.fit_predict(df[['age','income']])
y_pre

df['cluster']=y_pre
df

km.cluster_centers_
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['age'],df1['income'],color='g')
plt.scatter(df2['age'],df2['income'],color='b')
plt.scatter(df3['age'],df3['income'],color='r')

plt.xlabel('Age')
plt.xlabel('Income')
plt.legend()
```
